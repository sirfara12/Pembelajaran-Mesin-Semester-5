## Penjelasan Kegunaan Library

### 1. PyPREP
- Digunakan untuk preprocessing sinyal EEG (Electroencephalogram).  
- Fungsinya untuk membersihkan noise dan artefak (misalnya gangguan dari kedipan mata, kontraksi otot, atau alat perekam).  
- Cocok dipakai di bidang neurosains, Brain-Computer Interface (BCI), dan analisis aktivitas otak.  

---

### 2. SciPy
- Salah satu library ilmiah utama di Python.  
- Menyediakan banyak fungsi untuk matematika, statistik, optimasi, aljabar linear, dan pemrosesan sinyal.  
- Banyak digunakan untuk analisis data, filtering sinyal, transformasi Fourier, dan perhitungan numerik kompleks.  

---

### 3. Weights & Biases (wandb)
- Platform untuk tracking eksperimen Machine Learning dan Deep Learning.  
- Kegunaan utamanya:
  - Mencatat metrik training (loss, akurasi, dll).  
  - Menyimpan model, dataset, dan konfigurasi eksperimen.  
  - Membandingkan hasil eksperimen dengan mudah melalui dashboard interaktif.  
- Sangat membantu ketika melakukan eksperimen berulang agar prosesnya lebih terkontrol.  

---

### 4. pyECG
- Library Python untuk analisis sinyal ECG (Electrocardiogram).  
- Fungsinya:
  - Deteksi R-peak pada sinyal jantung.  
  - Ekstraksi fitur HRV (Heart Rate Variability).  
  - Analisis pola ECG untuk mendeteksi potensi masalah kesehatan jantung.  
- Biasanya dipakai di bidang biomedical engineering, kesehatan digital, dan penelitian medis.
-  tetapi pyECG sudah lama tidak diperbarui dan sulit dipasang di Python versi terbaru.  
  Sebagai pengganti, bisa digunakan neurokit2 atau ecg-detectors.

##### 2.Berikan contoh tentang dampak energi dan lingkungan terhadap pemanfaatan kecerdasan buatan dan bagaimana cara mengatasinya (Anda dapat memberikan pendapat pribadi). Sertakan referensi yang digunakan.

1. **Dampak Negatif**
- Pelatihan model AI skala besar (misalnya GPT, BERT, dan sejenisnya) membutuhkan energi listrik yang sangat besar, seringkali berasal dari sumber tidak terbarukan.
- Jejak karbon (carbon footprint) dari pelatihan model AI dapat setara dengan konsumsi energi ribuan rumah tangga.
- Pemanfaatan pusat data (data center) yang besar meningkatkan pemanasan global dan kebutuhan pendinginan intensif.


2. **Cara Mengatasinya**
- Menggunakan energi terbarukan (misalnya tenaga surya, angin) untuk pusat data AI.
- Menerapkan **model efisien** (misalnya model compression, knowledge distillation, transfer learning) agar tidak perlu melatih dari nol.
- Optimalisasi infrastruktur dengan green computing (server hemat energi, pendingin ramah lingkungan).
- Pendapat pribadi: Edukasi dan regulasi perlu diperkuat agar perusahaan AI wajib melaporkan jejak karbon serta mengalokasikan riset pada AI hijau (Green AI).


**Referensi:**
- Strubell, E., Ganesh, A., & McCallum, A. (2019). *Energy and Policy Considerations for Deep Learning in NLP*. ACL.
- Patterson, D., et al. (2021). *Carbon Emissions and Large Neural Network Training*. arXiv:2104.10350.
- Greenpeace. (2020). *Clicking Clean Report*.
